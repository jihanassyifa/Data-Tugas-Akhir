```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams['mathtext.fontset'] = 'cm'
plt.rcParams['font.family'] = 'STIXGeneral'
plt.rcParams.update({'font.size': 20})

df = pd.read_csv("D:\SATELITLAPAN_A1.txt", sep="\s+", header=None,
                  names=["No Katalog Satelit", "Epoch", "B*", "True Anomaly", "Height"])

df
```

```python
df['Epoch']
```

```python
fig, ax = plt.subplots(figsize=(19,6.5))
plt.title("Epoch Terhadap Parameter B* Satelit LAPAN-A1", fontsize=25)
plt.plot(df["Epoch"], df['B*'], color='black', linewidth=1.5)
plt.xlabel("Epoch (year)", fontsize=17)
plt.ylabel("Hambatan Atmosfer (B*)", fontsize=17)
plt.grid()
ax.text(0.01, 0.95, '(b)', horizontalalignment='left', verticalalignment='top', fontsize=20, transform=ax.transAxes)

plt.show()
```
