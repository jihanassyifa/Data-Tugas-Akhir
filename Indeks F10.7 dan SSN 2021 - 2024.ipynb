```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams['mathtext.fontset'] = 'cm'
plt.rcParams['font.family'] = 'STIXGeneral'
plt.rcParams.update({'font.size': 20})

df5 = pd.read_csv("D:\indeksantariksa.txt", sep="\s+", header=None,
                   names=["dates", "R", "F10.7 Index"],
                  parse_dates=['dates'])

df5
```

```python
date = df5['dates']
f_ind = df5['F10.7 Index']
ssn = df5['R']
```

```python
fig, ax = plt.subplots(2,figsize=(30,20),sharex=True)
fig.suptitle("Kondisi Cuaca Antariksa Tahun 2021 - 2024", fontsize=30, y=0.93)
ax[0].plot(date, f_ind, color='black', label="Indeks F10.7", linewidth=2)
#ax[0].set_xlabel("Tahun")
ax[0].set_ylabel("Indeks F10.7 (sfu)", fontsize=23)
ax[0].grid()
ax[0].text(0.01, 0.95, '(a)', horizontalalignment='left', verticalalignment='top', fontsize=40, transform=ax[0].transAxes)

ax[1].plot(date, ssn, color='black', label="R", linewidth=2)
ax[1].set_xlabel("Waktu", fontsize=23)
ax[1].set_ylabel("SSN", fontsize=23)
ax[1].grid()
ax[1].text(0.01, 0.95, '(b)', horizontalalignment='left', verticalalignment='top', fontsize=40, transform=ax[1].transAxes)

plt.show()

plt.savefig("antariksa.png",dpi=600)
```
