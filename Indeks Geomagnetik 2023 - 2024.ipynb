```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams['mathtext.fontset'] = 'cm'
plt.rcParams['font.family'] = 'STIXGeneral'
plt.rcParams.update({'font.size': 20})

df5 = pd.read_csv("D:\geomagindex.txt", sep="\s+", header=None,
                   names=["dates", "Kp Index", "R", "Dst Index", "F10.7 Index"],
                  parse_dates=['dates'])

df5
```

```python
date = df5['dates']
dst = df5['Dst Index']
kp = df5['Kp Index']
f_ind = df5['F10.7 Index']
ssn = df5['R']
```

```python
fig, ax = plt.subplots(2,figsize=(30,20),sharex=True)
fig.suptitle("Indeks Geomagnetik Tahun 2023 - 2024", fontsize=30, y=0.93)
ax[0].set_ylabel("Kp (*10)", fontsize=23)
ax[0].grid()
ax[0].text(0.01, 0.95, '(a)', horizontalalignment='left', verticalalignment='top', fontsize=40, transform=ax[0].transAxes)

ax[1].plot(date, dst, color='black', label="Dst", linewidth=1.5)
ax[1].set_xlabel("Waktu", fontsize=23)
ax[1].set_ylabel("Dst (nT)", fontsize=23)
ax[1].grid()
ax[1].text(0.01, 0.95, '(b)', horizontalalignment='left', verticalalignment='top', fontsize=40, transform=ax[1].transAxes)

# plt.title("Indeks Geomagnetik Kp Tahun 2023 - 2024")
# plt.plot(date, kp, color='black', label="Kp", linewidth=1.5)
# plt.xlabel("Tahun")
# plt.ylabel("Kp (*10)")
# plt.grid()
# ax.text(0.01, 0.95, '(a)', horizontalalignment='left', verticalalignment='top', fontsize=20, transform=ax.transAxes)

# fig, ax = plt.subplots(figsize=(19,7))
# plt.title("Indeks Geomagnetik Dst Tahun 2023 - 2024")
# plt.plot(date, dst, color='black', label="Dst", linewidth=1.5)
# plt.xlabel("Tahun")
# plt.ylabel("Dst (nT)")
# plt.grid()
# ax.text(0.01, 0.95, '(b)', horizontalalignment='left', verticalalignment='top', fontsize=20, transform=ax.transAxes)


# plt.figure(figsize =(20,5))
# plt.title("Indeks F10.7")
# plt.plot(date, f_ind, color='black', label="F10.7", linewidth=1.5)
# plt.xlabel("Tahun")
# plt.ylabel("Indeks F10.7")
# plt.grid()

# plt.figure(figsize =(20,5))
# plt.title("Sunspot Number")
# plt.plot(date, ssn, color='black', label="R", linewidth=1.5)
# plt.xlabel("Tahun")
# plt.ylabel("Sunspot Number")
# plt.grid()

# plt.show()
```
